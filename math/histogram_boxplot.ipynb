```python
from sklearn.datasets import fetch_california_housing
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = fetch_california_housing(as_frame=True)
df = data.frame
```

```python
df
```

```python
print(df['AveBedrms'].mean())
print(df['AveBedrms'].median())
print(df['AveBedrms'].mode())
print(df['AveBedrms'].std())
print(df['AveBedrms'].var())
print(df['AveBedrms'].quantile(0.25))
print(df['AveBedrms'].quantile(0.75))
print(df['AveBedrms'].quantile(0.75) - df['AveBedrms'].quantile(0.25))
print(df['AveBedrms'].describe())
print(df['AveBedrms'].describe()['75%'] - df['AveBedrms'].describe()['25%'])
```

```python
x = np.arange(0, 3 * np.pi, 0.1)
y = np.sin(x)

plt.plot(x, y)
```

```python
plt.plot(x, y, ':', lw=5, animated=True)
y_cos = np.cos(x)
plt.plot(x, y_cos, 'r--', lw=5, animated=True)
plt.xlabel('x')
plt.ylabel('y')
plt.title('sin(x)') 
plt.legend(['sin', 'cos'])
plt.show()
```
