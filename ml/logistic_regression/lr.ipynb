# Formular of Sigmoid(z), aka logistic function
$$
Sigmoid(z) = \frac{1}{1+e^{-z}}
$$

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

% matplotlib inline
```

```python
x = np.linspace(-10, 10, 100)
y = 1/(1+np.exp(-x))
plt.plot(x, y)
plt.show()
```
