```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
x=np.linspace(0,10,30)
y=np.sin(x)
plt.plot(x,y,'o',color='black')
```

```python
rng=np.random.RandomState(0)
for marker in ['o','.',',','x','+','v','^','<','>','s','d']:
    plt.plot(rng.rand(5),rng.rand(5),marker,label="marker='{0}'".format(market))
    
plt.legend(numpoints=1)
plt.xlim(0,1.8)
```

```python
plt.plot(x,y,'-s',color='black')
```

```python
plt.plot(x,y,'-.p',color='green',markersize=20,
        linewidth=6,markerfacecolor='black',
        markeredgecolor='red',markeredgewidth=6)
```

```python
plt.scatter(x,y,marker='o')
```

```python
rng=np.random.RandomState(0)
x=rng.randn(100)
y=rng.randn(100)
colors=rng.rand(100)
sizes=1000*rng.rand(100)

plt.scatter(x,y,c=colors,s=sizes,alpha=0.3,cmap='viridis')
plt.colorbar();
```
