```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.DataFrame({"age":[10,20,33,4,55]}, index = ["Daughter","Son","Uncle","Grandson","Dad"])
df.plot(kind="bar")
```

```python
df.plot.bar()
plt.xticks(rotation=-20, horizontalalignment="center")
plt.xlabel("Family Members")
plt.ylabel("Age")
plt.title("Family information")
```

```python
df.plot(kind="barh")
plt.ylabel("Family Members")
plt.xlabel("Age")
plt.title("Family information")
```

```python

df1 = pd.DataFrame({"2019":[89,23,44,54], "2020":[23,45,66,55], "2021": [33,66,77,44]}, index=["Bob","James","Ted","Mary"])
df1.plot.bar()
```

```python
df1.plot(kind='bar', stacked= True)
```

```python
df1[['2020','2019','2021']].plot(kind="bar", stacked=True)
```

```python
df1.apply(lambda x: x*100/sum(x), axis=1).plot(kind="bar", stacked=True)
```
