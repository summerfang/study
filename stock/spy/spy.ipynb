```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
spy_day = pd.read_csv('SPY_1day_sample.txt', header=None, parse_dates=True)
spy_day[0] = pd.to_datetime(spy_day[0], format='%Y-%m-%d')
spy_day[1] = spy_day[1].astype(float)
spy_day = spy_day.set_index(0)

spy_day.columns = ['open', 'high', 'low', 'close', 'volume']
plt.figure(figsize=(10, 5))
plt.plot( spy_day['high'])
plt.plot( spy_day['low'])
```

```python
spy_1imin = pd.read_csv('SPY_1min_sample.txt', header=None, parse_dates=True)
spy_1imin[0] = pd.to_datetime(spy_1imin[0], format='%Y-%m-%d %H:%M:%S')
spy_1imin[1] = spy_1imin[1].astype(float)
spy_1imin = spy_1imin.set_index(0)
spy_1imin.columns = ['open', 'high', 'low', 'close', 'volume']

plt.figure(figsize=(300, 50))
plt.plot( spy_1imin['high'])
plt.plot( spy_1imin['low'])
```
